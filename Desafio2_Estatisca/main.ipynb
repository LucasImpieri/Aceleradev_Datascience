# Desafio 2

## Objetivo

* Queremos conhecer melhor nossos clientes por estado. Para isso, iniciamos uma análise na pontuação de crédito. 
* Para realizar a verificação inicial, precisamos de alguns valores.
* Os valores são a média, a mediana, a moda e o desvio padrão da pontuação de crédito.

### Tópicos

* Neste desafio você aprenderá:

- Média;
- Mediana;
- Moda;
- Desvio padrão.

### Descrição dos dados:
- 'id': Identificador do cliente
- 'sobrenome': Sobrenome do cliente
- 'pontuacao_credito': Pontuação de crédito do cliente (quanto maior, melhor o cliente geralmente)
- 'estado_residencia': Estado de residência do cliente
- 'genero': Gênero do cliente
- 'nivel_estabilidade': Nível de estabilidade do cliente
- 'saldo_conta': Saldo disponível na conta do cliente
- 'numero_produtos': Número de produtos que o cliente consome
- 'possui_cartao_de_credito': Possui um cartão de crédito cadastrado
- 'membro_ativo': Membro acessa e consome frequentemente

```python
# agrupando estado_residencia com as estatisticas de pontuação de crédito 
df.groupby('estado_residencia')['pontuacao_credito'].describe()
```

```python
# agrupando estado com 'moda', 'mediana', 'media', 'desvio padrao' da pontuação de credito, 
# usando o numpy para fazer as operações, exceto para a moda (tem que usar uma serie do pandas)
# transpondo o DT para ficar igual ao arquivo de submissão (.T)
# renomeando os index para ficar igual ao arquivo (.rename)
# organizando as colunas 

aux = df.groupby('estado_residencia')['pontuacao_credito'].agg([pd.Series.mode, np.median, np.mean, np.std]).T.rename(index={'mode':'moda','median':'mediana','mean':'media','std':'desvio_padrao'})
aux = aux[['SC', 'RS', 'PR']]
aux
```

```python
# transformando o dataFrame em um arquivo JSON, agrupado por estado
aux.to_json('submissão.json', orient='columns')
```
